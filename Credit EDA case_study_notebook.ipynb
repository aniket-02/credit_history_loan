Treat the outliers using the inter quartile method.

```python
continuous_cols = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'REGION_POPULATION_RELATIVE', 'DAYS_BIRTH', 'DAYS_EMPLOYED']
for col in continuous_cols:
    Q1 = filtered_data[col].quantile(0.25)
    Q3 = filtered_data[col].quantile(0.75)
    IQR = Q3 - Q1
    Lower_Fence = Q1 - (1.5 * IQR)
    Upper_Fence = Q3 + (1.5 * IQR)
    filtered_data = filtered_data[~((filtered_data[col] < Lower_Fence) | (filtered_data[col] > Upper_Fence))]
```

```python
filtered_data['DAYS_BIRTH'] = filtered_data['DAYS_BIRTH'].abs()
filtered_data['DAYS_EMPLOYED'] = filtered_data['DAYS_EMPLOYED'].abs()
```

```python
#Check the boxplots to ensure outliers are treated.
for cols in continuous_cols:
    plt.figure(figsize=(12,8))
    sns.boxplot(filtered_data[cols])
```
